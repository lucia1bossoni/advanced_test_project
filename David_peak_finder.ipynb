```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = np.loadtxt("Data_David.txt")
length=(len(data))
x=[]
y1=[]
for i in range(length):
    temp=data[i]
    jaja=temp[0]
    x.append(jaja*1e-6)
#print(x)
for i in range(length):
    temp=data[i]
    jaja=temp[1]
    y1.append(jaja)

y2=[]
x2=[]
avg=5
for i in range(length-avg+1):
    temp_y=0
    temp_x=0
    for n in range(avg):
        temp_y=temp_y+y1[i+n]
        temp_x=temp_x+x[i+n]
    temp_y_new=temp_y/avg
    temp_x_new=temp_x/avg
    x2.append(temp_x_new)
    y2.append(temp_y_new)
#print(y1)
plt.xlabel("Voltage(V)")
plt.ylabel("Current(I)")
plt.plot(x,y1,x2,y2)
plt.yscale('log')
plt.show()
```
